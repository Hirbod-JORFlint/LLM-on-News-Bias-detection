```python
import pandas as pd
import matplotlib.pyplot as plt
```

First show what our data is made of

```python
SG_final1=pd.read_excel('/content/drive/MyDrive/final_labels_SG1.xlsx')
SG_final1.head()
```

```python
plt.bar(SG_final1['label_bias'].value_counts().index, SG_final1['label_bias'].value_counts().values)
plt.xlabel('Bias Plot')
plt.ylabel('Frequency')
plt.title('Label Bias Distribution for SG1')

plt.savefig("SG1_lb_bias")
plt.show()
```

```python
# Compute the count of outlets for each outlet
outlet = SG_final1['outlet'].value_counts()

# Sort in descending order
outlet = outlet.sort_values(ascending=False)

# Outlet Distribution for SG2
plt.bar(outlet.index, outlet.values)

plt.xlabel('Outlet Plot')
plt.ylabel('Frequency')
plt.title('Outlet Distribution for SG1')

# Rotate x-axis labels
plt.xticks(rotation='vertical')

plt.savefig("SG1_outlet")
plt.show()
```

```python

l_by_type = SG_final1.groupby('type')['label_bias'].value_counts()

# Plot the counts
l_by_type.plot(kind='bar')
plt.title('Count of Each Type based on political leaning')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Compute the count of label_bias for each outlet
l_by_outlet = SG_final1.groupby('outlet')['label_bias'].value_counts()

# Sort in descending order and select the top 10
l_by_outlet = l_by_outlet.sort_values(ascending=False).head(10)

# Plot the counts
l_by_outlet.plot(kind='bar')
plt.title('Count of Each Outlet-Type')
plt.xlabel('Outlet-Type')
plt.ylabel('Count')
plt.show()
```

```python
# Compute the count of label_bias for each topic
l_by_topic = SG_final1.groupby('topic')['label_bias'].value_counts()

# Sort in descending order and select the top 10
l_by_topic = l_by_topic.sort_values(ascending=False).head(10)

# Plot the counts
l_by_topic.plot(kind='bar')
plt.title('Count of Each Topic-Type')
plt.xlabel('Topic-Type')
plt.ylabel('Count')
plt.show()
```

```python
# Compute the count of label_opinion for each outlet
ol_by_outlet = SG_final1.groupby('outlet')['label_opinion'].value_counts()

# Sort in descending order and select the top 10
ol_by_outlet = ol_by_outlet.sort_values(ascending=False).head(10)

# Plot the counts
ol_by_outlet.plot(kind='bar')
plt.title('Count of Each Outlet-Opinion')

plt.xlabel('Outlet-Opinion')
plt.ylabel('Count')
plt.show()
```
